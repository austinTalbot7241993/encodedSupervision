```python
from math import log,sqrt
import numbers

import sys,os
sys.path.append('../Code/')
import numpy as np
from scipy import linalg as la
from scipy.special import gammaln
from scipy.sparse import issparse
from scipy.sparse.linalg import svds
from numpy import random as rand
from tqdm import trange
import matplotlib.pyplot as plt

from sklearn.utils import check_array, check_random_state
from sklearn.utils.extmath import randomized_svd,fast_logdet,svd_flip
from sklearn.utils.extmath import stable_cumsum
from sklearn.utils.validation import check_is_fitted
from sklearn import decomposition as dp

from datetime import datetime as dt
import os,sys,time
from scipy.io import savemat
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LR

from sPCA import spca_local,spca_enc

def csSim(x,y):
    return (np.dot(x,y)/(la.norm(x)*la.norm(y)))

def sortedEig(E):
    eigenValues,eigenVectors = la.eig(E)
    idx = eigenValues.argsort()[::-1]
    eigenValues = eigenValues[idx]
    eigenVectors = eigenVectors[:,idx]
    return eigenValues,eigenVectors

```

# Generating the data. Notice how the largest coefficient is unpredictive

```python
rand.seed(1993)
N = 100000 # Number of observations
p = 20 # Dimension of feature space
q = 1 # Supervised dimension
L = 3 # True latent dimension
eps_x = .5 # Noise for features
eps_y = .5 # Noise for supervision
S2 = rand.randn(N,L) # Latent factors
n_components = 2

#Generating the features with set eigenvalues
lamb1 = 5.
lamb2 = 3.0
lamb3 = 2.0

S2[:,0] = S2[:,0]*lamb1
S2[:,1] = S2[:,1]*lamb2
S2[:,2] = S2[:,2]*lamb3

W_orig = rand.randn(L,p)
W_orig[1] = W_orig[1] - np.dot(W_orig[0],W_orig[1])/np.dot(W_orig[0],W_orig[0])*W_orig[0]
W_orig[2] = W_orig[2] - np.dot(W_orig[0],W_orig[2])/np.dot(W_orig[0],W_orig[0])*W_orig[0]
W_orig[2] = W_orig[2] - np.dot(W_orig[1],W_orig[2])/np.dot(W_orig[1],W_orig[1])*W_orig[1]

W_orig[0] = W_orig[0]/la.norm(W_orig[0])
W_orig[1] = W_orig[1]/la.norm(W_orig[1])
W_orig[2] = W_orig[2]/la.norm(W_orig[2])

#Supervision strength
D_orig = rand.randn(L,q)
D_orig[0] = 0.0 #Largest coefficient unpredictive
D_orig[2] = D_orig[2]*2

#Actually generate the observations
X_nonoise = np.dot(S2,W_orig)
Y_nonoise = np.dot(S2,D_orig)
X2 = X_nonoise + eps_x*rand.randn(N,p)
Y2 = Y_nonoise + eps_y*rand.randn(N,q)


X2 = X2.astype(np.float32)
Y2 = Y2.astype(np.float32)
N = 50000
X = X2[:N]
Y = Y2[:N]
S = S2[:N]
X_true = X_nonoise[:N]
Y_true = Y_nonoise[:N]
P_true = np.dot(W_orig.T,W_orig)
```

## Go through a range of strengths in encoded model

```python
NN = 1000
mus = np.logspace(start=-1.5,stop=3,num=NN)
W_list = []
A_list = []
Strain_list = []
Stest_list = []
recon_train_losses = np.zeros(NN)
recon_test_losses = np.zeros(NN)
pred_train_losses = np.zeros(NN)
pred_test_losses = np.zeros(NN)
p_diffs = np.zeros(NN)
```

```python
for i in range(len(mus)):
    model = spca_enc(n_components=2,lamb = mus[i])
    model.fit(X,Y)
    W_enc = model.W_
    A_enc = model.A_
    S_tr = model.project(X)
    S_te = model.project(X2)
    X_recon_tr = model.reconstruct(X)
    X_recon_te = model.reconstruct(X2)
    Y_recon_tr = model.predict(X)
    Y_recon_te = model.predict(X2)
    Px = np.dot(W_enc,W_enc.T)
    p_diffs[i] = np.mean((Px-P_true)**2)
    recon_train_losses[i] = np.mean((X-X_recon_tr)**2)
    recon_test_losses[i] = np.mean((X2-X_recon_te)**2)
    pred_train_losses[i] = np.mean((Y_recon_tr-Y)**2)
    pred_test_losses[i] = np.mean((Y_recon_te-Y2)**2)
    W_list.append(W_enc)
    A_list.append(A_enc)
    Strain_list.append(S_tr)
    Stest_list.append(S_te)
    #print(mus[i],np.mean(X_recon_tr**2),np.mean(A_enc**2),np.mean(S_tr**2))
    
```

## Range of supervision strengths in local model

```python
W_list = []
A_list = []
Strain_list = []
Stest_list = []
recon_train_losses_l = np.zeros(NN)
recon_test_losses_l = np.zeros(NN)
pred_train_losses_l = np.zeros(NN)
pred_test_losses_l = np.zeros(NN)

trans_train_losses_l = np.zeros(NN)
trans_test_losses_l = np.zeros(NN)
trans_test_losses_x_l = np.zeros(NN)

p_diffs = np.zeros(NN)
s_drag = np.zeros(NN)
```

```python
for i in range(len(mus)):
    model = spca_local(n_components=2,lamb = mus[i])
    model.fit(X,Y)
    W_enc = model.W_
    D_enc = model.D_
    
    S_te = model.project(X2)
    S_tet = model.transform(X2,Y2)

    X_r = np.dot(S_tet,W_enc.T)
    Y_r = np.dot(S_tet,D_enc.T)
    
    X_p = np.dot(S_te,W_enc.T)
    Y_p = np.dot(S_te,D_enc.T)
    
    Px = np.dot(W_enc,W_enc.T)
    p_diffs[i] = np.mean((Px-P_true)**2)

    recon_test_losses_l[i] = np.mean((X_p-X2)**2)

    pred_test_losses_l[i] = np.mean((Y_p-Y2)**2)

    trans_test_losses_l[i] = np.mean((Y2-Y_r)**2)

    trans_test_losses_x_l[i] = np.mean((X_r-X2)**2)
    
    W_list.append(W_enc)
    Strain_list.append(S_tr)
    Stest_list.append(S_te)
    s_drag[i] = np.mean((S_te-S_tet)**2)
```

## Weird spike presumably due to shift in eigenspace

```python
fs1 = 20
fs2 = 22
lw = 2.5
lMus = np.log(mus)
fig, (ax1, ax2,ax3) = plt.subplots(1, 3);
fig.set_figheight(5)
fig.set_figwidth(15)
ax1.plot(lMus,trans_test_losses_x_l,c='r',linestyle='dashed',lw=lw,label='Local: Y known')
ax1.plot(lMus,recon_test_losses_l,c='k',linestyle='dotted',lw=lw,label='Local: Y unknown')
ax1.plot(lMus,recon_test_losses,c='g',linestyle='dashdot',lw=lw,label='Encoded')
ax1.set_xlabel('log($\mu$)',fontsize=fs1)
ax1.set_ylabel('MSE',fontsize=fs1)
ax1.legend(loc='upper left')
ax1.set_title('Reconstruction of X',fontsize=fs2)

ax2.plot(lMus,trans_test_losses_l,c='r',linestyle='dashed',lw=lw,label='Local: Y known')
ax2.plot(lMus,pred_test_losses_l,c='k',linestyle='dotted',lw=lw,label='Local: Y unknown')
ax2.plot(lMus,pred_test_losses,c='g',linestyle='dashdot',lw=lw,label='Encoded')
ax2.set_xlabel('log($\mu$)',fontsize=fs1)
ax2.set_ylabel('MSE',fontsize=fs1)
ax2.legend(loc='lower left')
ax2.set_title('Prediction of Y',fontsize=fs2)

ax3.plot(lMus,s_drag,lw=lw,c='k')
ax3.set_xlabel('log($\mu$)',fontsize=fs1)
ax3.set_ylabel('MSE',fontsize=fs1)
ax3.set_title('Local Factor Dragging',fontsize=fs2)
#plt.savefig('Example2.pdf')
```
